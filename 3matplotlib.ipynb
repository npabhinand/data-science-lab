```python
data={"java":22.2,"python":17.6,"php":8.8,"javascript":8,"C#":77,"C++":6.7}
course=list(data.keys())
value=list(data.values())
fig=plt.figure(figsize=(10,5))
plt.bar(course,value,color="black",width=.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=['g1','g2','g3','g4','g5']
girls=[25,32,30,35,29]
boys=[22,30,35,35,26]
x_axis=np.arange(len(x))
width=.4
plt.bar(x_axis -.2,boys,width,label="boys")
plt.bar(x_axis +.2,girls,width,label="girls")
plt.xticks(x_axis,x)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.array([22.2,17.6,8.8,8,77,6.7])
x=['java','python','php','javascript','c#','c++']
plt.pie(a, labels=x,autopct='%.1f')
plt.show()
```
